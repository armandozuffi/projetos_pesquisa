# Otimização de Perfil por Dephasing Lenght

```python
%matplotlib nbagg
import abel
import scipy
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt

from scipy.constants import c, m_e, e, epsilon_0
```

```python
#Parâmetros
#comprimento de onda d pulso
wlength_0 = 0.8e-6 # metros

#tempo do pulso
tau_0 = 50e-15 #seg

#freq. do pulso
omega_0 = (2*math.pi*c)/wlength_0

#densidade de elétrons
ne = np.linspace(1e16,1e19,100)

#laser strengh
a0 = 0.8
```

```python
# freq. da onda de plasma
omega_p = np.sqrt((ne*e*e*1e6)/(m_e*epsilon_0))

wlength_p = (2*math.pi*c)/omega_p
```

```python
#Dephasiing length Ld
Ld_1D = ((omega_0**2*wlength_p)/(omega_p**2))

Ld_2D = Ld_1D/2

Ld_3D = (4*omega_0**2*np.sqrt(a0)*c)/(3*omega_p**3)

#Pump depletion
Lpd = ((omega_0**2)*tau_0*c)/(a0**2 * omega_p**2)
```

```python
#Plot
plt.plot(ne,Ld_2D)
plt.plot(ne,Lpd)
plt.ylim(0,1.5)
plt.xlim(0,0.25e19)
```
